```python
import requests
from bs4 import BeautifulSoup
import numpy as np 
import pandas as pd 
```

```python
url="https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
response=requests.get(url)
response
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
article_head=[]
link=[]
author_name=[]
publish_data=[]
```

```python
a=soup.find_all('h2',class_="sc-1qrq3sd-1 MKjKb sc-1nmom32-0 sc-1nmom32-1 hqhUYH ebTA-dR")
for s in a :
    q=s.text
    article_head.append(q)
```

```python
c=soup.find_all('a',class_="sc-5smygv-0 nrDZj")
for f in c:
    k=f['href']
    link.append(k)
```

```python
v=soup.find_all('span',class_="sc-1thf9ly-2 bKddwo")
for y in v:
    f=y.text
    publish_data.append(f)
```

```python
g=soup.find_all('span',class_="sc-1w3fpd7-0 pgLAT")
for t in g:
    s=t.text
    author_name.append(s)
```

```python
## article_head=[] link=[] author_name=[] publish_data=[]

article_data=pd.DataFrame({'article_head':article_head,'author_name':author_name,'publish_data':publish_data,'links':link,})
```
