```python
import pandas as pd 
import numpy as np 
import requests
from bs4 import BeautifulSoup
```

```python
url='https://www.cnbc.com/world/?region=world'
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.content,'lxml')
news=soup.findAll('div',class_='LatestNews-headlineWrapper')
```

```python
head_name=[]
time=[]
links_main=[]

```

```python
for store in news:
    h=store.find('a',class_='LatestNews-headline').text
    head_name.append(h)
    
    fg=store.find('time',class_='LatestNews-timestamp').text
    time.append(fg)
    
     
```

```python
bsoup=BeautifulSoup(response.text,'html.parser')
```

```python
c=bsoup.find_all('a',attrs={'class':"LatestNews-headline"})
for l in c:
    link=l['href']
    links_main.append(link)
```

```python
data_news=pd.DataFrame({'headline':head_name,'time_of_news':time,'web_links':links_main})
```
